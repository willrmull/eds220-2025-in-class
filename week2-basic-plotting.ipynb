```python
file_path = 'data/wetlands_seasonal_bird_diversity.csv'

df = pd.read_csv(file_path)
```

```python
# Plot a graph of the spring bird surveys at Mugu Lagoon 
# with respect to the years. Include some basic customization.
df.plot(x = 'year', 
        y = ['MUL_spring'],
        title = 'Spring Bird Surveys at Mugu Lagoon',
        ylabel='Number of bird species',
        color='green',
        legend=False)
```

```python
# Read in data
URL = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/main/inst/extdata/penguins.csv'
penguins = pd.read_csv(URL)

penguins.head()
```

```python
# Select the bill_length_mm and bill_depth_mm columns in the penguins dataframe 
# and then update the kind parameter to box to make boxplots of the bill length and bill depth.
penguins[['bill_length_mm', 'bill_depth_mm']].plot(kind='box',
                                title='Penguin flipper lengths',
                                xlabel='Flipper length (mm)',
                                grid=True)
```
